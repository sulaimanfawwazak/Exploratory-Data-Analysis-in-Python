## **Feature Relationships**
* Scatterplot

```python
df.plot(kind='scatter', x='Speed_mph', y='Height_ft', title='Coaster Speed vs Height')
plt.show()
```

```python
ax = sns.scatterplot(data=df, x='Speed_mph', y='Height_ft', hue='Year_Introduced')
ax.set_title('Coaster Speed vs Height')
plt.show()
```

```python
sns.pairplot(
  data=df, 
  vars=['Year_Introduced', 'Speed_mph', 'Height_ft', 'Inversions', 'Gforce'], 
  hue='Type_Main'
  )
plt.show()
```

```python
df_corr = df[['Year_Introduced', 'Speed_mph', 'Height_ft', 'Inversions', 'Gforce']] \
  .dropna() \
  .corr()
df_corr
```

```python
sns.heatmap(df_corr, annot=True)
```

## **Step 5: Ask Questions about The Data**
Try to answer a specific question about the data using a plot or statistic

Example:
* What are the locations with the fastes roller coasters, with minimum of 10 roller coasters?

```python
ax = df.query('Location != "Other"') \
  .groupby('Location')['Speed_mph'] \
  .agg(['mean', 'count']) \
  .query('count >= 10') \
  .sort_values('mean')['mean'] \
  .plot(kind='barh', figsize=(12, 5), title='Average Coaster Speed by Location')
ax.set_xlabel('Average Coaster Speed')
plt.show()
```
